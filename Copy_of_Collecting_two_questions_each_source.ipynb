```python
#Import the dataset

from google.colab import files
import pandas as pd

# Upload the file from your local machine
uploaded = files.upload()

# Assuming the file is named 'input_updated.csv'
df = pd.read_csv('cleaned_ai_responses_tone.csv')
print(df.head())
```

```python
import random

# Group data by 'source' (AI channel)
grouped = df.groupby('Source')

# Create a dictionary to store 2 posts for each AI tool
sampled_posts = {}

# For each AI channel, sample 2 posts
for source, group in grouped:
    # Ensure there are at least 2 posts from this AI tool
    if len(group) >= 2:
        # Randomly sample 2 posts
        sampled_posts[source] = group['Cleaned AI Response'].sample(2).tolist()
    else:
        # If less than 2 posts, take all available posts
        sampled_posts[source] = group['Cleaned AI Response'].tolist()

# Print the sampled posts for each AI tool
for tool, posts in sampled_posts.items():
    print(f"\nPosts generated by {tool}:")
    for idx, post in enumerate(posts, 1):
        print(f"{idx}. {post}")
```
